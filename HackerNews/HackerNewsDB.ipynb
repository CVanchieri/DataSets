```
# load the csv files.
db1 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb1.csv')
db2 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb2.csv')
db3 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb3.csv')
db4 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb4.csv')
db9 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb9.csv')
db10 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb10.csv')
db11 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb11.csv')
db12 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb12.csv')
db13 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb13.csv')
db14 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb14.csv')
db15 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb15.csv')
db16 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb16.csv')
db17 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb17.csv')
db18 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb18.csv')
db19 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb19.csv')
db20 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb20.csv')
db21 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb21.csv')
db22 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb22.csv')
db23 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb23.csv')
db24 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb24.csv')
db25 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb25.csv')
db26 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb26.csv')
db27 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb27.csv')
db28 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb28.csv')
db29 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb29.csv')
db30 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb30.csv')
db31 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb31.csv')
db32 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb32.csv')
db33 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb33.csv')
db34 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb34.csv')
db35 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb35.csv')
db36 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb36.csv')
db37 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb37.csv')
db38 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb38.csv')
db39 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb39.csv')
db40 = pd.read_csv('https://raw.githubusercontent.com/CVanchieri/DataSets/master/HackerNews/commentsdb40.csv')
# merge the csv files.
HNfinalDB = pd.concat([db1, db2, db3, db4, db9, db10, db11, db12, db13, db14, db15, db16, 
                       db17, db18, db19, db20, db21, db22, db23, db24, db25, db26, db27, 
                       db28, db29, db30, db31, db32, db33, db34, db35, db36, db37, db38, 
                       db39, db40], sort=True)
# drop unecessary columns.
HNfinalDB = HNfinalDB.drop(columns=['dead', 'deleted', 'parent', 'time'])
# use datetime on time_ts column.
HNfinalDB['time_ts'] = pd.to_datetime(HNfinalDB['time_ts'], infer_datetime_format=True)
# seperate the date and time from time_ts column.
HNfinalDB['date'] = [d.date() for d in HNfinalDB['time_ts']]
HNfinalDB['time'] = [d.time() for d in HNfinalDB['time_ts']]
# sepereate the year, month, day for date column.
HNfinalDB['year'] = HNfinalDB['date'].map(lambda x: x.year)
HNfinalDB['month'] = HNfinalDB['date'].map(lambda x: x.month)
HNfinalDB['day'] = HNfinalDB['date'].map(lambda x: x.day)
# drop the time_ts column.
HNfinalDB = HNfinalDB.drop(columns=['time_ts', 'date'])
# reorganize the columns.
HNfinalDB = HNfinalDB[['year', 'month', 'day', 'time', 'author', 'id', 'text', 'ranking']]
# drop any duplicate columns based on id column.
HNfinalDB.drop_duplicates(subset ="id", keep = "first", inplace = True)
# show data frame shape.
print(HNfinalDB.shape)
# show the data frame with headers.
HNfinalDB.head(10)
```

```
# authors with the most comments in the database.
HNtop = HNfinalDB['author'].value_counts()[:10].sort_values(ascending=False)
print("--authors with most comments--")
print(HNtop)
```
